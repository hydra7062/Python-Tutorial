String Indexing

```python
# we access the value through indexing

# ex-->            apple
# indexing --->    a[0], p[1], p[2],l[3],e[4]
# number of elements -->  5


string = "elephant"

print(string[0])



# using Ranging


print(string[0 :5])
print(string[2 :5])


# string     [initial/starting : ending : every nth element ]

print(string[2 :5:2])



# Negative indexing

# negative indexing start from right hand side
#  elephant

# t = -1
# n =-2
# a= -3
# h = -4
# and so on

print(string[-3])
```

Slicing

```python
a = " Slicing "

print(a[1:4])

print(a[0:-4])
print(a[0:-3])
print(a[0:-2])
```

Reverse a string

```python
# 1. using slicing

s = "String"

print(s[: : -1])
```

Deleting and Updating

```python
# Delete

# Deleting using Slicing

string = "my name is jarvis"

string2 = string[10 : ]

print(string2)


string3 = string[0:2]
print(string3)



# delete entire string

str1 = "string"
del str1

print(str1)
```

Formatting

```python
string = " {} {} {} ".format('apple', 'banana','mango')
print(string)

# Positional Formatting
string1 = " {2} {0} {1} ".format('B for Boy','C for Cat','A for Apple' )
print(string1)

# keyword Formatting

string2 = " {A} {B} {C} ".format(B = 'B for Boy',C = 'C for Cat',A = 'A for Apple' )
print(string2)
```

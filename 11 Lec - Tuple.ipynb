Access Tuple Items
using indexing

```python
My_tuple = ("Mango", "Banana", "Apple")

print(My_tuple[0])

print(My_tuple[0:2])

print(My_tuple[-2:])
```

Updation in Tuple

```python
t = ("rohan", 'pooja', " Sushant", "Ajay ")

# for updation we convert into list than after tuple

t1 = list(t)

t1[0]= "Salman Khan"

t = tuple(t1)
print(t)


```

append

```python
t = ("rohan", 'pooja', " Sushant", "Ajay ")

t1 = list(t)

t1.append("Shahrukh Khan")

t = tuple(t1)
print(t)
```

Adding 2 tuple

```python
x = ("Apple", " Bnana", " Orange")

y = ("Potato", "Ginger")

print(x+y)
```

Remove items

```python
x = ("Apple", " Bnana", " Orange")
y = list(x)

y.remove("Apple")

x= tuple(y)
print(x)
```
